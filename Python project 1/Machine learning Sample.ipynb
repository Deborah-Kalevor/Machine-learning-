```python
# Removing non interor color types values 
df = df[~df['Interior_Color'].isin(['Select', '.'])]
df.Interior_Color.unique()
```

```python
# Removing non fuel types values 
non_fuel_values = ['5-Speed A/T', 'A/T', '6-Speed M/T', 'CVT Transmission']
df = df[~df['Fuel_Type'].isin(non_fuel_values)]
df.Fuel_Type.unique()
```

```python
# Removing non Transmission types values 
df = df[~df['Transmission'].isin(['â€“', 'NOT SPECIFIED', '2'])]
df.Transmission.unique()
```

```python
# Filling in missing values with their average
df["Comfort_Rating"].fillna(value=df["Comfort_Rating"].mean(), inplace=True)
df["Performance_Rating"].fillna(value=df["Performance_Rating"].mean(), inplace=True)
```

```python
# covert Price to a float 
# Convert object column to float using astype()
# replace 'Not Priced' with NaN
df['Price'] = df['Price'].replace('Not Priced', np.nan)

# drop NaN values since the empty values are few 
df = df.dropna(subset=['Price'])

# convert Price to a float
df['Price'] = df['Price'].replace('[\$,]', '', regex=True).astype(float)
```

```python
df.Make.unique()
df.Year.unique()
df.Model.unique()
df.Condition.unique()
```

```python
# factorizing string colums
# Make
Car = {}
code = 1
for Make in df.Make.unique():
    Car[Make] = code
    code+=1
 
df.replace(Car, inplace=True)
#Years
Years= {}
code = 1
for Year in df.Year.unique():
    Years[Year] = code
    code+=1
    
 
df.replace(Years, inplace=True)
#Models
Models= {}
code = 1
for Model in df.Model.unique():
    Models[Model] = code
    code+=1
    
 
df.replace(Models, inplace=True)

#Conditions
Conditions= {}
code = 1
for Condition in df.Condition.unique():
    Conditions[Condition] = code
    code+=1
    
 
df.replace(Conditions, inplace=True)
```

```python
df.head()
```

```python
df.Exterior_Color.unique()
df.Interior_Color.unique()
```

```python
#Exterior Color
Exterior_Colors= {}
code = 1
for Exterior_Color in df.Exterior_Color.unique():
    Exterior_Colors[Exterior_Color] = code
    code+=1
    
 
df.replace(Exterior_Colors, inplace=True)

# Interior Color
Interior_Colors= {}
code = 1
for Interior_Color in df.Interior_Color.unique():
    Interior_Colors[Interior_Color] = code
    code+=1
# since the Exterior and interior have similar values to avoid the same color factor in both
# we used this formulae
df['Exterior_Color'] = df['Exterior_Color'].replace(Exterior_Colors)
df['Interior_Color'] = df['Interior_Color'].replace(Interior_Colors)

```

```python
df.head()
```

```python
df.Fuel_Type.unique()
df.Transmission.unique()
df.Engine.unique()
```

```python
#Fuel Types
Fuel_Types= {}
code = 1
for Fuel_Type in df.Fuel_Type.unique():
    Fuel_Types[Fuel_Type] = code
    code+=1
    
 
df.replace(Fuel_Types, inplace=True)

#Transmissions
Transmissions= {}
code = 1
for Transmission in df.Transmission.unique():
    Transmissions[Transmission] = code
    code+=1
    
 
df.replace(Transmissions, inplace=True)

#Engines
Engines= {}
code = 1
for Engine in df.Engine.unique():
    Engines[Engine] = code
    code+=1
    
 
df.replace(Engines, inplace=True)
```

```python
# splitting the data sets into training and testing 
                      
columns = ["Year","Make","Model","Condition","Exterior_Color","Interior_Color","Fuel_Type","Transmission" ,"Engine", 
           "Comfort_Rating","Performance_Rating"]   
x = df[ columns ] 
y = df.Price
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size= 0.2)
```

```python
np.random.seed(42)
from sklearn.linear_model import LinearRegression
model1 = LinearRegression()
```

```python
model1.fit (x_train,y_train)
```

```python
predicted = model1.predict( x_test)
```

```python
y.head()
```

```python
#Accuracy of the data using the mean squared error
from sklearn.metrics import mean_squared_error
error = mean_squared_error(y_test,predicted)
print(error)
```

```python
#root of the mean square
np.sqrt(error)
```

```python
#Accuracy of the Linear Regression using the R-square score
from sklearn.metrics import r2_score
# Calculate the R-squared score
r2 = r2_score(y_test,predicted)
```

```python
# Print the R-squared score
print('R-squared:', r2)
```
